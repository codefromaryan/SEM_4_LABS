```python
# Driving Permission 20240802159

age = int(input("Enter your age: "))
if age < 18:
  print("You are not eligible")
else:
  dp = input("Do you have driving license?: ")

  if age >= 18 and dp == "yes":
    print("You have driving permission")
  elif age >= 18 and dp == "no":
    print("You do not have driving permission")
  else:
    print("Something went wrong! Please use yes or no and only numbers for age")
```
