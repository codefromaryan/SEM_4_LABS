```python
#Basic Pattern Matching Functions

# ✅ re.match()
# Matches pattern at the beginning of the string
import re
pattern = r"hello"
text = "hello world"
result = re.match(pattern, text)
if result:
 print("Match found")
```

```python
# ✅ re.search()
# Searches the pattern anywhere in the string
result = re.search(r"world", "hello world")
print(result.group())
```

```python
# ✅ re.findall()
# Returns all matches as a list
text = "cat bat rat"
result = re.findall(r".at", text)
print(result) # ['cat', 'bat', 'rat']
```

```python
#✅ re.sub()
# Replaces pattern with another string
text = "I like apples"
result = re.sub(r"apples", "oranges", text)
print(result) # I like oranges
```

```python
# Validation Examples
# Email Validation
import re
email = "test@gmail.com"
pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"
if re.match(pattern, email):
 print("Valid Email")
else:
 print("Invalid Email")
```

```python
#Phone Number Validation (10 digits)
phone = "9876543210"
pattern = r"^\d{10}$"
if re.match(pattern, phone):
 print("Valid Phone Number")
```

```python
#Password Validation
#● At least 8 characters
#● Contains uppercase, lowercase, digit
password = "Abc12345"
pattern = r"^(?=.*[a-z])(?=.*[A-Z])(?=.*\d).{8,}$"
if re.match(pattern, password):
 print("Strong Password")
```

```python
import re

def validate_email(email):
    pattern = r"^[\w\.-]+@[\w\.-]+\.\w+$"
    return re.match(pattern, email)

def validate_phone(phone):
    pattern = r"^\d{10}$"
    return re.match(pattern, phone)

def validate_password(password):
    pattern = r"^(?=.*[a-z])(?=.*[A-Z])(?=.*\d).{8,}$"
    return re.match(pattern, password)

def find_words(text):
    return re.findall(r"\b\w+\b", text)

def replace_word(text):
    return re.sub(r"bad", "good", text)

def search_word(text):
    result = re.search(r"python", text, re.IGNORECASE)
    return result.group() if result else None


while True:
    print("\n--- REGEX APPLICATION MENU ---")
    print("1. Validate Email")
    print("2. Validate Phone Number")
    print("3. Validate Password")
    print("4. Find Words in Text")
    print("5. Replace Word in Text")
    print("6. Search Word in Text")
    print("7. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        email = input("Enter email: ")
        if validate_email(email):
            print("✅ Valid Email")
        else:
            print("❌ Invalid Email")

    elif choice == "2":
        phone = input("Enter phone number: ")
        if validate_phone(phone):
            print("✅ Valid Phone Number")
        else:
            print("❌ Invalid Phone Number")

    elif choice == "3":
        password = input("Enter password: ")
        if validate_password(password):
            print("✅ Strong Password")
        else:
            print("❌ Weak Password")

    elif choice == "4":
        text = input("Enter text: ")
        words = find_words(text)
        print("Words found:", words)

    elif choice == "5":
        text = input("Enter text: ")
        new_text = replace_word(text)
        print("Updated text:", new_text)

    elif choice == "6":
        text = input("Enter text: ")
        word = search_word(text)
        if word:
            print("Found word:", word)
        else:
            print("Word not found")

    elif choice == "7":
        print("Exiting program...")
        break

    else:
        print("Invalid choice! Try again.")
```
