```python
#Example: Single Inheritance
class Animal:
  def speak(self):
    print("Animal makes a sound")
class Dog(Animal): # Inheriting from Animal
  def bark(self):
    print("Dog barks")
# Object creation
d = Dog()
d.speak() # Inherited method
d.bark()
```

```python
#Example: Method Overriding
class Animal:
 def speak(self):
  print("Animal speaks")
class Cat(Animal):
 def speak(self): # Overriding parent method
  print("Cat meows")
c = Cat()
c.speak()
```

```python
#Using super()
class Animal:
 def __init__(self):
  print("Animal Constructor")
class Dog(Animal):
 def __init__(self):
  super().__init__() # Call parent constructor
 print("Dog Constructor")
d = Dog()
```

```python
#Example: Multilevel Inheritance
class Animal:
 def eat(self):
  print("Animal eats food")
class Dog(Animal): # Inherits from Animal
 def bark(self):
  print("Dog barks")
class Puppy(Dog): # Inherits from Dog (which already inherits Animal)
 def weep(self):
   print("Puppy weeps")
# Object creation
p = Puppy()
p.eat() # From Animal
p.bark() # From Dog
p.weep() # From Puppy
```
