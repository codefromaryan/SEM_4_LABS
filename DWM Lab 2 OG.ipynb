```python
  import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/indian_student_dirty_dataset_realistic (3).csv')
df.head()
```

##1) Replacing Blanks with Nan

```python
print("Initial Dataset Preview")
df.info()

print("\n Dataset Info")
df.info()

print("\n Missing Values (Before Cleaning)")
display(df.isnull().sum())

df.head
```

##2) Handle Missing Values

```python
# Numerical columns
num_cols = ["Age", "CGPA", "Attendance_Percentage",
            "Annual_Fees", "Year_of_Study", "Admission_Year"]

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)

# Categorical columns
cat_cols = df.select_dtypes(include="object").columns

for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

##3) Remove Noicy / Invalid Data

```python
# Age should be between 17 and 25
df.loc[(df["Age"] < 17) | (df["Age"] > 25), "Age"] = df["Age"].median()

# CGPA should be between 0 and 10
df.loc[(df["CGPA"] < 0) | (df["CGPA"] > 10), "CGPA"] = df["CGPA"].median()

# Attendance should be between 0 and 100
df.loc[(df["Attendance_Percentage"] < 0) |
       (df["Attendance_Percentage"] > 100),
       "Attendance_Percentage"] = df["Attendance_Percentage"].median()

# Admission Year should be between 2015 and 2025
df.loc[(df["Admission_Year"] < 2015) |
       (df["Admission_Year"] > 2025),
       "Admission_Year"] = df["Admission_Year"].median()

# Annual Fees should not be negative
df.loc[df["Annual_Fees"] < 0, "Annual_Fees"] = df["Annual_Fees"].median()
```

##4) Fix Inconsistent Data

```python
# Gender standardization
df["Gender"] = df["Gender"].str.lower().replace({
    "m": "Male",
    "male": "Male",
    "f": "Female",
    "female": "Female"
})

# City and State formatting
df["City"] = df["City"].str.title()
df["State"] = df["State"].str.title()

# Fix state short form
df["State"] = df["State"].replace({"Up": "Uttar Pradesh"})
```

##5) Clean Course Column


```python
df["Course"] = df["Course"].str.strip().replace({
    "Btech": "B.Tech",
    "BSc": "B.Sc",
    "mba": "MBA"
})
```

##6️) Fix Department Column


```python
df["Department"] = df["Department"].replace({
    "AIDS": "AI&DS"
})
```

##7️) Caste Category Formatting

```python
df["Caste_Category"] = df["Caste_Category"].str.upper()
```

##8️) Standardize Binary Columns

```python
binary_map = {
    "Y": "Yes", "y": "Yes", "yes": "Yes",
    "N": "No",  "n": "No",  "no": "No"
}

df["Scholarship_Status"] = df["Scholarship_Status"].replace(binary_map)
df["Hostel_Resident"] = df["Hostel_Resident"].replace(binary_map)
```
