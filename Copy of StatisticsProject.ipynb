```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
file_path='/content/drive/MyDrive/PHD/Statistics/bestsellers with categories.csv'
df=pd.read_csv(file_path)
df.head()
df.shape
```

```python
length=df.shape[0]
print('length of Given dataset',length)
```

```python
df.head()
```

```python
df['Genre'].replace({'Non Fiction': 0, 'Fiction': 1}, inplace=True)
df.head()
```

```python
#Finding min , max, range of column User Rating, Reviews,Price
column_name='User Rating'
#Finding max without using function
data=df[column_name]
data=data.sort_values(ascending=False)
max=data.iloc[0]
print('Max of User Rating',max)

#Finding min without using function
data=df[column_name]
data=data.sort_values(ascending=True)
min=data.iloc[0]
print('Min of User Rating',min)
```

```python
#Range without function for column User Rating
range=max-min
print('Range of User Rating',range)

#Range with Function
review_range=df['Reviews'].max()-df['Reviews'].min()
print('Range of Reviews',review_range)

price_range=df['Price'].max()-df['Price'].min()
print('Range of price',price_range)
```

```python
#Arithmetic Mean
#With inbuilt function
mean_user_rating=df['User Rating'].mean()
print('Mean of User Rating',mean_user_rating)

mean_reviews=df['Reviews'].mean()
print('Mean of Reviews',mean_reviews)

mean_price=df['Price'].mean()
print('Mean of Price',mean_price)

#Without inbuilt function
sum=0
for i in df['User Rating']:
  sum=sum+i
mean_user_rating=sum/length
print('Mean of User Rating',mean_user_rating)
```

```python
import math
product = 1
for num in df['User Rating']:
    product *= num
#print(product)
geometric_mean =  (float)(math.pow(product, (1 / len(df['User Rating']))))

print('Geometric mean for price column ',geometric_mean)
sum_reciprocals = 0
for num in df['User Rating']:
    if num != 0:
      sum_reciprocals += 1 / num

harmonic_mean = len(df['User Rating']) / sum_reciprocals

print('Harmonic mean for price column ',harmonic_mean)
```

```python
#Median
print('Median of User Rating',df['User Rating'].median())
print('Median of Reviews',df['Reviews'].median())
print('Median of Price',df['Price'].median())
```

```python
#Mode
print('Mode of User Rating',df['User Rating'].mode())
print('Mode of Reviews',df['Reviews'].mode())
print('Mode of Price',df['Price'].mode())
```

```python
#variance
print('Variance of User Rating',df['User Rating'].var())
print('Variance of Reviews',df['Reviews'].var())
print('Variance of Price',df['Price'].var())
#StandardDeviation
print('Standard Deviation of User Rating',df['User Rating'].std())
print('Standard Deviation of Reviews',df['Reviews'].std())
print('Standard Deviation of Price',df['Price'].std())
#MeanDeviation
mean_dev = abs(df['User Rating'] - df['User Rating'].mean()).mean()
print('Mean Deviation of User Rating',mean_dev)
print('Mean Deviation of Reviews',df['Reviews'].apply(abs).mean())
print('Mean Deviation of Price',df['Price'].apply(abs).mean())
```

```python
#Quartile
print('Quartile of User Rating',df['User Rating'].quantile([0.25,0.5,0.75]))
print('Quartile of Reviews',df['Reviews'].quantile([0.25,0.5,0.75]))
print('Quartile of Price',df['Price'].quantile([0.25,0.5,0.75]))
```

```python
df['User Rating'].plot(kind='hist', bins=10,color=['orange', 'red'], edgecolor='black', title='Histogram of User Ratings Data')
plt.xlabel('User Ratings values')
plt.ylabel('Nos of book ')
plt.show()
```

```python
sns.lineplot(y=df['User Rating'],x=df['Year'],hue=df['Genre']);
plt.ylabel("Ratings")
plt.xlabel("Years");
```

```python
MaxPriceBookList=df.groupby('Name')['Price'].max()
MaxPriceBookList=MaxPriceBookList.sort_values(ascending=False).head(10)
sns.barplot(x=MaxPriceBookList,y=MaxPriceBookList.index)
plt.xlabel("Price")
plt.ylabel("Top Price Books")
plt.title('Bar Chart for Top Price Book');
```

```python
genre_average=df.groupby(['Genre'])['Price'].mean()
sns.barplot(x=genre_average.index,y=genre_average);
plt.xlabel("Genre: NonFiction=0 Fiction=1")
plt.ylabel("Average Price")
plt.title('Bar Chart for Average Price of Genre');
```

```python
Gen_count = df['Genre'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(Gen_count, labels=['Non Fiction', 'Fiction'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
plt.title('Distribution of Non Fiction vs. Fiction Books')
plt.show()
```

```python
#Skewness
User_Rating_Skewness=df['User Rating'].skew()
Reviews_Skewness=df['Reviews'].skew()
Price_Skewness=df['Price'].skew()

if User_Rating_Skewness>0:
  print('User Rating has Left Skew')
elif User_Rating_Skewness<0:
  print('User Rating has Right Skew')
else:
  print('User Rating has Normal Distribution')

if Reviews_Skewness>0:
  print('Reviews has Left Skew')
elif Reviews_Skewness<0:
  print('Reviews has Right Skew')
else:
  print('Reviews has Normal Distribution')

if Price_Skewness>0:
    print('Price has Left Skew')
elif Price_Skewness<0:
    print('Price has Right Skew')
else:
    print('Price has Normal Distribution')
```

```python
#Visualise Skewness
sns.histplot(df['Reviews'], kde=True)
plt.title(f'Data Distribution (Skewness: {User_Rating_Skewness:.2f})')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```
