##Func of square

```python
def square(x):
  return x*x

for i in range(3):
  print(f"The square of {i} is {square(i)}")
```

##Area of rectangle

```python
def area(length, width):
  return length * width

dimensions = [(5, 4), (10, 3)]

for l, w in dimensions:
  rect_area = area(l, w)
  print(f"The area of a rectangle with length {l} and width {w} is {rect_area}")
```

##Add 10 nums to any given list

```python
def add_ten_numbers(input_list):
  for i in range(1, 11):
    input_list.append(i)
  return input_list

my_list = [10, 20, 30]
updated_list = add_ten_numbers(my_list)
print(f"Original list: {my_list}")
print(f"List after adding 10 numbers: {updated_list}")

another_list = []
updated_another_list = add_ten_numbers(another_list)
print(f"Original empty list: {[]}")
print(f"Empty list after adding 10 numbers: {updated_another_list}")
```

## Print add of all elements of list

```python
def sum_list_elements(input_list):
  total_sum = 0
  for item in input_list:
    total_sum += item
  return total_sum

# Example usage:
my_numbers = [1, 2, 3, 4, 5]
list_sum = sum_list_elements(my_numbers)
print(f"The sum of elements in {my_numbers} is: {list_sum}")

another_numbers = [10, 20, 30, 40]
list_sum_2 = sum_list_elements(another_numbers)
print(f"The sum of elements in {another_numbers} is: {list_sum_2}")
```

## Print power of any number

```python
def power(base, exponent):
  return base ** exponent

# Example usage:
print(f"2 raised to the power of 3 is: {power(2, 3)}")
```

## Print given number is even or odd

```python
is_even_or_odd = lambda num: "Even" if num % 2 == 0 else "Odd"

# Example usage:
number1 = 10
print(f"The number {number1} is {is_even_or_odd(number1)}")

number2 = 7
print(f"The number {number2} is {is_even_or_odd(number2)}")
```

## Make function for factorial

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)

# Example usage:
number = 5
print(f"The factorial of {number} is: {factorial(number)}")

number = 0
print(f"The factorial of {number} is: {factorial(number)}")
```
