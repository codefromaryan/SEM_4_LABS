```python
import pandas as pd
import numpy as np

df = pd.read_csv("student_data_1000_rows.csv") # While using Jupiter Notebook use 'r' before the double colon.
df.head()
```

```python
print("Dataset Shape: ", df.shape)
print("\nDataset Info: \n")
df.info()
```

```python
def classify_data_type(series):
    if series.dtype == 'object':
        return "Qualitative"
    elif np.issubdtype(series.dtype, np.integer):
        return "Quantitative - Discrete"
    elif np.issubdtype(series.dtype, np.floating):
        return "Quantitative - Continuous"
    else:
        return "Unknown"
```

```python
def Identify_scale(column):
    col = column.lower()

    if col in ["name","marks"]:
        if col == "marks":
            return "Ordinal"
        return "Nominal"

    if col in ["sub1", "sub2", "sub3", "sub4","sub5"]:
        return "Interval"


    if col in ["study_hr_per_week", "sgpa","sr_no"]:
        return "Ratio"

    return "Ratio (Assumed)"
```

```python
results = []

for col in df.columns:
    data_type = classify_data_type(df[col])
    scale = Identify_scale(col)
    results.append([col,data_type,scale])


classification_df = pd.DataFrame(results,columns=["attribute","Data Types","Scale of Measurment"])
classification_df
```
