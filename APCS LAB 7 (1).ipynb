```python
# Lab 7: File Handling and Exception Handling in Python

# ----------- FILE HANDLING PART -----------
# Name: Aryan Chavan
# Prn: 20240802159

def file_operations():
    filename = "example.txt"

    try:
        # Writing to a file
        with open(filename, "w") as file:
            file.write("Hello, Python File Handling!\n")
        print("File written successfully.")

        # Appending to the file
        with open(filename, "a") as file:
            file.write("Adding a new line.\n")
        print("Data appended successfully.")

        # Reading entire file
        with open(filename, "r") as file:
            content = file.read()
            print("\nFull File Content:")
            print(content)

        # Reading line by line
        print("Reading line by line:")
        with open(filename, "r") as file:
            for line in file:
                print(line.strip())

    except FileNotFoundError:
        print("File not found.")
    except IOError:
        print("File operation failed.")
    finally:
        print("File handling operations completed.\n")
```

```python
# ----------- EXCEPTION HANDLING PART -----------
# Name: Aryan Chavan
# Prn: 20240802159

def exception_examples():
    # Example 1: Division by zero
    try:
        a = 10
        b = 0
        result = a / b
        print(result)
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")

    # Example 2: Invalid input
    try:
        num = int("abc")
    except ValueError:
        print("Error: Invalid conversion.")

    # Example 3: Using else and finally
    try:
        num = int("10")
    except ValueError:
        print("Conversion failed.")
    else:
        print("Conversion successful:", num)
    finally:
        print("Execution completed.\n")


# ----------- RAISING EXCEPTION -----------
# Name: Aryan Chavan
# Prn: 20240802159

def check_age(age):
    if age < 18:
        raise ValueError("Age must be 18 or above")
    return "Eligible"


# ----------- MAIN FUNCTION -----------

def main():
    print("---- FILE HANDLING ----")
    file_operations()

    print("---- EXCEPTION HANDLING ----")
    exception_examples()

    print("---- CUSTOM EXCEPTION ----")
    try:
        age = int(input("Enter your age: "))
        print(check_age(age))
    except ValueError as e:
        print("Error:", e)


# Run the program
if __name__ == "__main__":
    main()
```
