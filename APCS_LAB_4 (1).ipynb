```python
# ===============================
# LIST
# ===============================
my_list = [10, 20, 30, 40]

my_list.append(50)
my_list.remove(20)

print("List:", my_list)
print("First element:", my_list[0])

# ===============================
# TUPLE
# ===============================
my_tuple = (1, 2, 3, 4)

print("Tuple:", my_tuple)
print("Second element:", my_tuple[1])

# ===============================
# SET
# ===============================
my_set = {1, 2, 3, 3, 4}

my_set.add(5)
my_set.remove(2)

print("Set:", my_set)

# ===============================
# DICTIONARY
# ===============================
my_dict = {
    "name": "Jishan",
    "age": 20,
    "course": "B.Tech"
}

my_dict["age"] = 21
my_dict["city"] = "Pune"

print("Dictionary:", my_dict)
print("Name:", my_dict["name"])

# ===============================
# EXTRA OPERATIONS
# ===============================

# Loop through dictionary
for key, value in my_dict.items():
    print(key, ":", value)
```
