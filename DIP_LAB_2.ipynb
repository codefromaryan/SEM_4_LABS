##**Aim : To study an implement histogram processing techniques and to compare their effects on image enhancement.**


## Setup and Imports

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.exposure import match_histograms
```

##**TOOK A DARK IMAGE FROM  SPLASH.COM**

## Loading and Displaying the RGB Image

```python
img = cv2.imread('/content/premium_photo-1769019242610-1fae526340cc.avif')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

print("Grayscale Image Shape: ", img.shape)

plt.imshow(img)
plt.title('Acquired image')
plt.axis('off')
```

## Grayscale Conversion

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.figure(figsize = (6,6))
plt.imshow(img_gray, cmap = 'gray')
plt.title('Image divided into grid(Pixel Representation)')
```

## Pixel Level Analysis

```python
pixel_block = img_gray[100:110, 100:110]

print("10x10 pixel matrix: \n")
print(pixel_block)
```

```python
plt.imshow(pixel_block, cmap='gray')
plt.title('10x10 pixel matrix grid')
plt.colorbar()
plt.axis('off')
```

## Grayscale Histogram Equalization

```python
equalized_img = cv2.equalizeHist(img_gray)

plt.figure(figsize = (10,4))

plt.subplot(1,2,1)
plt.imshow(img_gray, cmap='gray')
plt.title('Original grayscale')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(equalized_img, cmap='gray')
plt.title('Equalized grayscale')
plt.axis('off')
```

```python
plt.figure(figsize = (12,4))

#BEFORE
plt.subplot(1,2,1)
plt.hist(img_gray.ravel(), bins = 256)
plt.title("Grayscale Histogram (Before Equalization)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

#AFTER
plt.subplot(1,2,2)
plt.hist(equalized_img.ravel(), bins = 256)
plt.title("Grayscale Histogram (Before Equalization)")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

## RGB Channel Processing and Histogram Equalization

```python
r, g, b = cv2.split(img)

plt.figure(figsize = (15,8))

# RED CHANNEL
plt.subplot(3,2,1)
plt.imshow(r, cmap = 'gray')
plt.title("Red Channel Image")
plt.axis('off')

plt.subplot(3,2,2)
plt.hist(r.ravel(), bins = 256)
plt.title("Red Channel Histogram")

# GREEN CHANNEL
plt.subplot(3,2,3)
plt.imshow(g, cmap = 'gray')
plt.title("Green Channel Image")
plt.axis('off')

plt.subplot(3,2,4)
plt.hist(g.ravel(), bins = 256)
plt.title("Green Channel Histogram")

#blue channel
plt.subplot(3,2,5)
plt.imshow(b, cmap = 'gray')
plt.title("Blue Channel Image")
plt.axis('off')

plt.subplot(3,2,6)
plt.hist(b.ravel(), bins = 256)
plt.title("Blue Channel Histogram")
```

```python
r_eq = cv2.equalizeHist(r)
g_eq = cv2.equalizeHist(g)
b_eq = cv2.equalizeHist(b)

plt.figure(figsize = (15,8))

# RED CHANNEL
plt.subplot(3,2,1)
plt.imshow(r_eq, cmap = 'gray')
plt.title("Equalized Red Channel Image")
plt.axis('off')

plt.subplot(3,2,2)
plt.hist(r_eq.ravel(), bins = 256)
plt.title("Equalized Red Channel Histogram")

# GREEN CHANNEL
plt.subplot(3,2,3)
plt.imshow(g_eq, cmap = 'gray')
plt.title("Equalized Green Channel Image")
plt.axis('off')

plt.subplot(3,2,4)
plt.hist(g_eq.ravel(), bins = 256)
plt.title("Equalized Green Channel Histogram")

#blue channel
plt.subplot(3,2,5)
plt.imshow(b_eq, cmap = 'gray')
plt.title(" Equalized Blue Channel Image")
plt.axis('off')

plt.subplot(3,2,6)
plt.hist(b_eq.ravel(), bins = 256)
plt.title("Equalized Blue Channel Histogram")
```
