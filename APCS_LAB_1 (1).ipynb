##The Fibonacci Sequence

```python
#20240802159

def generate_fibonacci(n):
 a, b = 0, 1
 series = []
 for _ in range(n):
  series.append(a)
  a, b = b, a + b
 return series
print(generate_fibonacci(10))
```

## Reversing a Number

```python
#20240802159

def reverse_number(n):
 reversed_num = 0
 while n > 0:
  digit = n % 10
  reversed_num = (reversed_num * 10) + digit
  n //= 10
 return reversed_num
print(reverse_number(5678))
```
