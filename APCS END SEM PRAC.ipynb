```python
def fibonacci(n):
    if n <= 0: return 0   # base condition
    if n == 1: return 1   # base condition
    return fibonacci(n-1) + fibonacci(n-2)
for i in range(10):
    print(fibonacci(i))
```

```python
class Student:
    # Class variable (shared by all)
    college = "D Y Patil"
    # Constructor — called automatically
    def __init__(self, name, age, marks):
        self.name  = name       # instance variable
        self.age   = age
        self.marks = marks
    def display(self):
        print(f"Name:    {self.name}")
        print(f"Age:     {self.age}")
        print(f"College: {Student.college}")
    def grade(self):
        if   self.marks >= 90: return "O"
        elif self.marks >= 75: return "A"
        elif self.marks >= 60: return "B"
        else:
          return "C"

s1 = Student("Rahul", 20, 88)
s2 = Student("Priya", 21, 95)
s1.display()
s2.display()
print(s1.grade())
print(s2.grade())
```

```python
# Single Inheritance
class Animal:
    def __init__(self, name):
        self.name = name
    def eat(self):
        print(f"{self.name} is eating")
class Dog(Animal):
    def bark(self):
        print(f"{self.name}: Woof!")
d = Dog("Rex")
d.eat()     # inherited
d.bark()    # own
```

```python
# Multilevel Inheritance
class Vehicle:
    def start(self): print("Vehicle started")
class Car(Vehicle):
    def drive(self): print("Car driving")
class SportsCar(Car):
    def turbo(self): print("Turbo boost!")
sc = SportsCar()
sc.start()  # from Vehicle
sc.drive()  # from Car
sc.turbo()  # own
```

```python
# Multiple Inheritance
class Father:
    def skills(self): print("Cricket, Cooking")
class Mother:
    def skills(self): print("Teaching, Art")
class Child( Mother, Father):  # MRO: Father first
    def my_skills(self): print("Coding, Music")
c = Child()
c.skills()      # from Father (MRO)
c.my_skills()
```

```python
class BankAccount:
  def __init__(self, name, balance):
    self.name = name
    self.balance = balance

  def deposit(self, amount):
    self.balance += amount
    print(f"{self.name} deposited {amount}. New balance: {self.balance}")

  def withdraw(self, amount):
    if self.balance >= amount:
      self.balance -= amount
      print(f"{self.name} withdrew {amount}. Remaining balance: {self.balance}")
    else:
      print(f"Insufficient funds for {self.name}. Current balance: {self.balance}")

class SavingAccount (BankAccount):
  def __init__(self, name, balance, interest_rate):
    super().__init__(name, balance)
    self.interest_rate = interest_rate

  def display_interest_rate(self):
    print(f"The interest rate for {self.name}'s savings account is {self.interest_rate}%")

s = SavingAccount("rahul", 30, 5.3)
s.deposit(100)
s.withdraw(20)
s.display_interest_rate()
```

```python
class Student:
  college = "D.Y.Patil International Uni"

  def __init__(self, name, age, marks):
    self.name = name
    self.age = age
    self.marks = marks

  def display(self):
    print(f"Name: {self.name}")
    print(f"Age: {self.age}")
    print(f"Marks: {self.marks}")

  def grades(self):
    if self.marks >= 90:
      print("A+")
    elif self.marks >= 80:
      print("A")
    elif self.marks >= 65:
      print("B")
    else:
      print("F")

s1 = Student("Rahul", 19, 85)

s1.display()
s1.grades()
```

```python
class Vehicle:
  def start(self):
    print("Engine Starts")

class Forward(Vehicle):
  def move(self):
    print("Moves forward.")

class Stop(Forward):
  def stop(self):
    print("Car stops.")

v = Stop()
v.start()
v.move()
v.stop()
```
